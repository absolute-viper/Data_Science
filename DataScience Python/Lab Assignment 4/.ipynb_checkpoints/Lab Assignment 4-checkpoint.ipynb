```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/ritik/Jupyter Notebooks/DataSets/AirQualityUCI.csv",sep=';',decimal=',')
df.drop(['Unnamed: 16','Unnamed: 15'],axis=1,inplace=True)
df.dropna(how='any',inplace=True)
for i in range(len(df)):
     df['Time'][i]=df['Time'][i].replace('.',':')
df['Date']=pd.to_datetime(df['Date'])
df
```

```python
pd_year=df['Date'].dt.year
pd_month=df['Date'].dt.month
df['Year']=pd_year
df['Month']=pd_month
group_2004 = df.groupby('Year')
df_2004 = group_2004.get_group(2004)
group_2004 = df_2004.groupby('Month',as_index=False)
median_2004 = group_2004.agg({'CO(GT)':['median']})
```

```python
#### QUES 1
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.plot(median_2004['Month'],median_2004['CO(GT)'])
```

```python
#### QUES 2
plt.style.use('ggplot')
plt.title('Monthly Median CO Concentration In 2004')
plt.grid(True)
plt.ylabel("Sensor Response")
plt.xlabel("Month")
plt.plot(median_2004['Month'],median_2004['CO(GT)'])
```
