```python
#### QUES 32/33/34
def Count_Noise(data,series):
    details = data.apply(lambda x : True
        if x[series] == -200.000000 else False, axis = 1)
    num_rows = len(details[details == True].index)
    return num_rows

def Noise_Iterate():
    for i in range(3,12):
        Column = df.columns[i]
        Noise = Count_Noise(df,Column)
        Percent = Noise/df.shape[0]*100
        out = ""
        if Percent > 10:
            df.drop(Column,axis=1,inplace=True)
            out = "Dropped"
        print("{0}, Noise: {1}, Percentage: {2} {3}".format(Column,Noise,Percent,out))
Noise_Iterate()
```
