```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/ritik/Jupyter Notebooks/DataSets/company_sales_data.csv")
df
```

```python
###### Ques 1
plt.plot(df['month_number'],df['total_profit'])
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.show()
```

```python
####### Ques 2
plt.plot(df['month_number'],df['total_profit'],color='red',linestyle='dotted',linewidth='3',marker='o',label='Profit')
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.legend(loc='lower right')
plt.show()
```

```python
####### Ques 3
x = df['month_number']
plt.plot(x, df['facecream'],label='Face Cream')
plt.plot(x, df['facewash'],label='Face Wash')
plt.plot(x, df['toothpaste'],label='Toothpaste')
plt.plot(x, df['bathingsoap'],label='Bathing Soap')
plt.plot(x, df['shampoo'],label='Shampoo')
plt.plot(x, df['moisturizer'],label='Moisturizer')
plt.xlabel('Month Number')
plt.ylabel('Total Sales')
plt.legend(loc='upper left')
plt.show()
```

```python
####### Ques 4
x = df['month_number']
plt.scatter(x,df['toothpaste'],label='toothpaste')
plt.grid(linestyle='-')
plt.xlabel("Month Number")
plt.ylabel("Total Sales")
plt.legend(loc='upper left')
plt.show()
```

```python
####### Ques 5
x = df['month_number']
width = 0.4
plt.bar(x-width/2, df['facecream'],width, label='Face Cream')
plt.bar(x+width/2, df['facewash'],width, label='Face Wash')
plt.xlabel("Month Number")
plt.ylabel("Total Sales")
plt.legend(loc='upper left')
plt.show()
```

```python
def create_bins(lower_bound, width, quantity):
    bins = []
    for low in range(lower_bound, lower_bound + quantity*width + 1, width):
        bins.append((low, low+width))
    return bins
```
